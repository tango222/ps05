### 2 Decision Boundary
Your next task is to plot the decision boundary when using logistic regression. You will also play a little
with feature engineering.
If you are uncertain about what is decision boundary, I recommend to consult James et al. (2015) book
Section 2. For instance, Figure 2.13 on p38 depicts a 2D classification case where certain X1, X2 values are
classified as orange and others as blue. Decision boundary is the dashed winding line that separates these
two regions.
There are two broad strategies to plot it. In any case, you have first to estimate (train) your model.
Thereafter you have to predict the classes (cancer/no cancer here) on a regular dense grid that covers the
parameter space (this is the small blue/orange dots on figure 2.13). Afterwards you can either plot your
predicted values with a certain color code, or alternatively, say, set predicted M to 1 and predicted B to
0, and make a contour plot for a single contour at level 0.5. You may also combine these both methods.
Some example code in R is provided here, python version will be in a separate file.
We ignore training/testing/overfitting issues for now.


### 2.1 kNN Case
First, let's explore the decision boundary using k-NN.
Pick two features. I recommend to use a few that show relative strong correlation with diagnosis,

Note: if your features are of very different scale, you should either scale these into a roughly equal

2. Plot the actual data and the decision boundary on the same plot. Ensure that actual observations

Here is an example code in R on how you may do this. We just generate some random data and use
logistic regression.

using concpoints.worst, and area.worst

```python
#scale data from columns and take sample of 100 rows
scaler = MinMaxScaler()
wdbc_s = (wdbc_sample[['diagnosis','concpoints.worst', 'area.worst']]).sample(100)
wdbc_scale = scaler.fit_transform(wdbc_s[['concpoints.worst', 'area.worst']])

#making each column it's own variable to be easier later
concw = pd.Series(data = wdbc_scale[:, 0])
areaw = pd.Series(wdbc_scale[:, 1])
diag = np.asarray(wdbc_s['diagnosis'])
```

```python
#plot for knn

m = KNeighborsClassifier(n_neighbors = 4).fit(wdbc_scale, wdbc_s['diagnosis'])
## Now create the regular grid
ex1 = np.linspace(concw.min(), concw.max(), 100)
ex2 = np.linspace(areaw.min(), areaw.max(), 100)
xx1, xx2 = np.meshgrid(ex1, ex2)

# unlike R-s 'expand.grid', meshgrid creates matrices
g = np.stack((xx1.ravel(), xx2.ravel()), axis=1)
# we create the design matrix by stacking the xx1, xx2
# after unraveling those into columns
## predict on the grid
hatY = m.predict(g).reshape(100,100)
# imshow wants a mtrix, so we reshape the predicted vector
#into one
plt.imshow(hatY, extent=(concw.min(), concw.max(), areaw.min(), areaw.max()),
interpolation='none', origin='lower',
# you need to specify that the image begins from below,
# not above, otherwise it will be flipped around
alpha=0.3)
plt.scatter(concw, areaw, c=diag)
plt.show()
```

### 3 ROC curve: which estimator is the best
Finally, let's see which estimator is the best. We compare these in two ways: first using accuracy, precision,
recall; and thereafter with ROC curve. We stay with the two features you have been using so far, for the
logistic one you also add the engineered features. Consult James et al. (2015, p 148) for what is the ROC
curve. Now we are going to compare the best models of k-NN, logistic regression, and SVM.
1. split your data into training-validation parts (say, 75-25).
2. select good models from all of your previous sections (i.e. k-NN, logistic, SVM). You may select all
models you tried, but if some of those were clearly inferior feel free to skip those.
3. Train each model on training data. Using the validation data for prediction and present the confusion
matrix and compute accuracy, precision, recall, and F-score.
Note: in this point you have to implement these algorithms yourself! do not use existing libraries!
Now let's compare these models with ROC curves by playing with different thresholds.
4. Use the models you used above for computing the accuracy and all that. Now instead of predicting
the oucome class, predict probability that the outcome belongs to a given class (say, Malignant).
The library you use may do it already anyway.
Note that while Logistic regression and SVM normally predict probability of the target class, at
least internally, this may not be the case for k-NN. If the software you are using supports it, use the
proportion of the successes among your k nearest neighbors as the probability (well, it only makes
sense if k > 1).
5. Now pick a sequence of probability thresholds between 0 and 1 (for instance, 0, 0.1, 0.2, ...). For
each model and each threshold, treat your predictions to be Malignant if it it's probability is at
least as big as the threshold.
6. Based on these predictions, compute false positive rate and true positive rate for each threshold
value. Plot these rates for each threshold and each model. This is your ROC curve.


```python
#split data into training and validation parts for knn and svm
wdbc_X = wdbc_data[['concpoints.worst', 'area.worst']]
X_train, X_test, y_train, y_test = train_test_split(
    wdbc_X, wdbc_data.diagnosis, test_size = 0.25, random_state = 11)
```

```python
(kpredict == 1).sum()
```

```python
#knn part
kreg = KNeighborsClassifier(n_neighbors = 4).fit(X_train, y_train)
kpredict = kreg.predict(X_test)

#computing accuracy 
kacc = ((kpredict == y_test) == True).sum()/len(kpredict)
print(kacc)

#computing recall
krec = ((kpredict == 1) & (y_test == 1)).sum()/(y_test == 1).sum()
print(krec)
#computing precision
kpre = ((kpredict == 1) & (y_test == 1)).sum()/(kpredict == 1).sum()
print(kpre)
#fscore
f_knn = 2/((1/krec) + (1/kpre))
print(f_knn)
```

```python
#knn confusion matrix
confusion_matrix(y_test, kpredict)
```

```python
#svm part with kernel 'rbf'
sclf = SVC(kernel = 'linear', probability=True).fit(X_train, y_train)
spredict = sclf.predict(X_test)

#computing accuracy 
sacc = ((spredict == y_test) == True).sum()/len(spredict)
print(sacc)
#computing recall
srec = ((spredict == 1) & (y_test == 1)).sum()/(y_test == 1).sum()
print(srec)

#computing precision
spre = ((spredict == 1) & (y_test ==1)).sum()/(spredict == 1).sum()
print(spre)
#fscore
f_svc = 2/((1/srec) + (1/spre))
print(f_svc)
```

```python
#ROC curve

#change training and validation data back for knn and svm
wdbc_X = wdbc_data[['concpoints.worst', 'area.worst']]
X_train, X_test, y_train, y_test = train_test_split(
    wdbc_X, wdbc_data.diagnosis, test_size = 0.25, random_state = 11)

#knn prediction probability
kprob_pred = kclf.predict_proba(X_test)[:, 1]
ktable = {'pred': kpredict, 'true' : y_test, 'prob' : kprob_pred}
ktable = pd.DataFrame(data = ktable)

#svm prediction probability 
sprob_pred = sclf.predict_proba(X_test)[:, 1]
stable = {'pred': spredict, 'true' : y_test, 'prob' : sprob_pred}
stable = pd.DataFrame(data = stable)

#changing train and val data to log data
wdbc_X['cwsq'] = np.square(wdbc_X['concpoints.worst'])
wdbc_X['awsq'] = np.square(wdbc_X['area.worst'])
wdbc_X['ctimesa'] = wdbc_X['concpoints.worst'] * np.square(wdbc_X['area.worst'])
X_train, X_test, y_train, y_test = train_test_split(
    wdbc_X, wdbc_data.diagnosis, test_size = 0.25, random_state = 11)

#log prediction probability 
lprob_pred = lreg.predict_proba(X_test)[:, 1]
ltable = {'pred': lpredict, 'true' : y_test, 'prob' : lprob_pred}
ltable = pd.DataFrame(data = ltable)
```

```python
def makeROC(threshes ,table):
    fpr = []
    tpr = []
    counts = []
    for thresh in threshes: 
        t = table.loc[(table.prob <= thresh)]
        tp = ((t.true == 1) & (t.pred == 1)).sum()/((t.true == 1).sum())
        fp = ((t.pred == 1) & (t.true == 0)).sum()/((t.true == 0).sum())
        tpr.append(tp)
        fpr.append(fp)
        counts.append(len(t))
    return tpr, fpr, counts
```

```python
kROC = pd.DataFrame(columns = ['tpr', 'fpr'])
kROC['tpr'], kROC['fpr'], counts = makeROC(np.arange(0.1, 1.1, 0.1),ktable)
```
